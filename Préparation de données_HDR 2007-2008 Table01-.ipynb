```python
import pandas as pd
import numpy as np

df=pd.read_excel("HDR 2007-2008 Table 01.xlsx")
df.head(12)
```

```python
df=pd.read_excel("HDR 2007-2008 Table 01.xlsx", header=6)
df=df.dropna(how="all",axis=1)
df=df.dropna(how="all", axis=0).reset_index(drop=True)
df.head(6)
```

```python
columns = []
for c in df.columns:
       columns.append([str(v) for v in df[c][:2] if pd.notnull(v)])
columns
```

```python
columns=["HDI rank","country","Human development index (HDI) value\n2005",
         "Life expectancy at birth\n(years)\n2005",
         "Reference 1",
         "Adult literacy rate\n(% aged 15 and above)\n1995-2005",
         "Reference 2",
         "Combined gross enrolment ratio for primary, secondary and tertiary education\n(%)\n2005",
         "Reference 3",
         "GDP per capita\n(PPP US$)\n2005",
         "Reference 4",
         "Life expectancy index",
         "Education index",
         "GDP index",
         "GDP per capita (PPP US$) rank minus HDI rank"
]
df.columns=columns
df.drop([0,1], inplace=True)
df = df.reset_index(drop=True)
df.head()
```

```python
hdi_categories = ["HIGH HUMAN DEVELOPMENT", "MEDIUM HUMAN DEVELOPMENT", "LOW HUMAN DEVELOPMENT"]
df= df.drop(df[df["HDI rank"].isin(hdi_categories)].index, axis=0).reset_index(drop=True)
```

```python
df.tail(30)
```

```python
df2=df.drop(np.arange(177,len(df)), axis=0).reset_index(drop=True)
```

```python
df2.loc[:, "HDI category"] = np.where(df2["HDI rank"] > 155,
                                     "LOW HUMAN DEVELOPMENT",
                                     np.where(df2["HDI rank"] > 70,
                                              "MEDIUM HUMAN DEVELOPMENT",
                                              "HIGH HUMAN DEVELOPMENT"
                                             )
                                    )
df2.sort_values(by="HDI rank", inplace=True)
df2.head()
```
